```python
## Find average battery life in 'bateria' column from each id in the df 
statistics = df.groupby('id')['bateria'].mean().reset_index()
statistics.columns = ['ID', 'Average Battery']


##  group 'cobertura' column and get the mean of each id , append to statistics df 
coverage = df.groupby('id')['cobertura'].mean().reset_index()
coverage.columns = ['ID', 'Average Coverage']


## get duplicate value of total_distance column of each'id' 

total_distance = df.groupby('id')['total_distance'].apply(lambda x: x.unique()[0]).reset_index()

total_distance.columns = ['ID', 'Total Distance']
print(total_distance)

## Merge statistics and coverage df  by 'ID' column
statistics= pd.merge(statistics, coverage, on='ID')

## Merge statistics and total_distance df  by 'ID' column
statistics= pd.merge(statistics, total_distance, on='ID')

# ## export to csv 
statistics.to_csv('statistics.csv', index=False)
```
